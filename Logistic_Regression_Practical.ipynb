```python
# Taking out the petal width column
X = pd.DataFrame(iris['data'][:,3])
```

```python
X
```

```python
iris['target'] == 2
```

```python
# converting the True and False into 0 and 1
# 1 ---> virginica
# 0  --> not virginica
```

```python
y = (iris['target'] == 2).astype(int)
y
```

```python
X
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
X_test
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()  # object creation
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
# importing the accuracy score parameter
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```
